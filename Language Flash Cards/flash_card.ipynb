```python
from tkinter import *
import pandas as pd
from random import choice
```

```python
BACKGROUND_COLOR = "#B1DDC6"
timer = None
french_word = None
english_word = None
#canvas_back_id = None
```

```python
#------------------------------------------Fetching the data--------------------------------------------------------------
try:
    df1 = pd.read_csv("data/words_to_learn.csv")
    words_df = df1.to_dict(orient="records")
    words_tolearn = words_df
except FileNotFoundError:
    df1 = pd.read_csv("data/french_words.csv")
    words_df = df1.to_dict(orient="records")
    words_tolearn = words_df

def next_word():
    next_dict = choice(words_df)
    global french_word,english_word,flip_timer
    window.after_cancel(flip_timer)
    french_word = next_dict["French"]
    english_word = next_dict["English"]
    
    canvas.itemconfig(card_background, image=card_front_img)
    canvas.itemconfig(lang_text,fill="black",text="French")
    canvas.itemconfig(word_text,fill="black",text=french_word)
    flip_timer = window.after(3000,show_word)
    

def show_word():
    global english_word
    #To change the image:
    canvas.itemconfig(card_background, image=card_back_img) 
    canvas.itemconfig(lang_text,fill="white",text="English")
    canvas.itemconfig(word_text,fill="white",text=english_word)

def right_func():
    save_word()
    next_word()
    
def wrong_func():
    next_word()

def save_word():
    words_tolearn.remove({"French":french_word,"English":english_word})
    words_tolearn_df = pd.DataFrame(words_tolearn)
    words_tolearn_df.to_csv("data/words_to_learn.csv", index=False)

    
    
```

```python
#------------------------------------------Creating UI--------------------------------------------------------------
window = Tk()
#window.geometry("1000x800")
window.title("Flashy")
window.config(padx=50,pady=50,bg=BACKGROUND_COLOR)
flip_timer = window.after(3000,show_word)

#--Flash Card--)
canvas = Canvas(width=800,height=526,bg=BACKGROUND_COLOR,highlightthickness=0)

#Back card
card_back_img = PhotoImage(file="images/card_back.png")
cc_back_img = canvas.create_image(400,270,image=card_back_img)

#Front card

card_front_img = PhotoImage(file="images/card_front.png")
card_background = canvas.create_image(400,270,image=card_front_img)
lang_text = canvas.create_text(400,150,text="French",fill="black",font=("Ariel",40,"italic"))
word_text = canvas.create_text(400,263,text="word",fill="black",font=("Ariel",60,"bold"))
canvas.grid(column=0,row=0,columnspan=2)




#--Add ❌ Button--
wrong_image = PhotoImage(file="images/wrong.png")
wrong_button = Button(image=wrong_image,bg=BACKGROUND_COLOR,highlightthickness=0,borderwidth=0,command=wrong_func)
wrong_button.grid(column=0,row=1,padx=50)


#--Add ✔ Button--
right_image = PhotoImage(file="images/right.png")
right_button = Button(image=right_image,bg=BACKGROUND_COLOR,highlightthickness=0,borderwidth=0,command=right_func)
right_button.grid(column=1,row=1,padx=50,pady=50)



next_word()

window.mainloop()
```
